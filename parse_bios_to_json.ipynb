```python
import json
import os
```

```python
files = os.listdir('bios/')
```

```python
max_len = 0
max_file = ''

full_bio = {}

for file in files:
    with open(f'bios/{file}', 'r',  encoding="utf-8") as f:
        bio = f.read()
  
    split = bio.split('\n')
    
    data = {}
    data['name'] = split[0]

    for el in split[1:]:

        if ('cm,' in el) and ('kg)' in el):
            physicall = el.split('(')[-1].split(',')
            data['psysicall'] = {'weight': physicall[1][:-1], 'heigh': physicall[0]}
            continue

        for small_el in el.split(' ▪ '):
            basic = small_el.split(': ')
            try:
                data[basic[0]] = basic[1]
            except IndexError:
                if split[0] != basic[0]:
                    data['nick'] = basic[0]
                
    full_bio[len(full_bio)] = data
```
