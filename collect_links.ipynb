```python
from selenium import webdriver
import pandas as pd
from tqdm import tqdm
```

```python
driver = webdriver.Chrome(executable_path='chromedriver.exe')
```

```python
driver.get('https://www.basketball-reference.com/leagues/NBA_2022_per_game.html')
```

```python
rows = driver.find_elements_by_class_name('full_table')
```

```python
names = []
links = []
for row in tqdm(rows):
    links.append(row.find_element_by_tag_name('a').get_attribute('href'))
    names.append(row.find_element_by_tag_name('a').text)
    
df = pd.DataFrame({'name': names, 'link': links})
```
